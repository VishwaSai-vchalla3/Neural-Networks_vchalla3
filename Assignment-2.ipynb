```python
import tensorflow as tf
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Import or Load the data from the .csv file..
data=genfromtxt("glass.data.csv",delimiter=',')
print(data)
print("DataType: ",type(data))
```

```python
#Prints the dimensions of the glass dataset
print("Shape of DataSet is: ",data.shape)
```

```python
#From the above data observation we can remove indexing column..
#Matrix consist of i/p and o/p where Last column is considered to be output..
X=data[:,1:10]#Shape of X matrix is 1 to 9[Which Consists of Input Features Alone.]
print("Shape Of X: ",X.shape)
```

```python
#y is the last column..
y=data[:,10]
print(y)#Printing the y data..
print("Shape Of y is : ",y.shape)
```

```python
#First 5 rows of the X Values Are Shown Below..
print("First 5 Rows are : ",X[:5])
```

```python
#Printing the Unique values and its Count from y.
#s=set(y)
#print("Unique Values Of Output(y) are : ",s)#Prints Unique Values.
#print("Count of Unique Values: ",len(s))#Count of unique Values
print("Unique Values Of 'y' are : ",np.unique(y))
print("Number Of Unique Values are: ",len(np.unique(y)))
```

```python
#Summary Statistics Of X features
print("Mean: ",np.mean(X,axis=0))
print("Standard Deviation: ",np.std(X,axis=0))
print("Maximum Value: ",np.max(X,axis=0))
print("Minimum Value: ",np.min(X,axis=0))
```

```python
#Plot a histogram for each of the features.

#Visualization
fig,axs = plt.subplots(4,3,figsize=(15,15))

#Plotting the histogram for each feature
counter=0
for i in range(4):
    for j in range(3):
        counter = counter+1
        if counter>X.shape[1]:break
        axs[i,j].hist(X[:,i],bins=20,color='blue',alpha=0.7)
        axs[i,j].set_title(f'Feature {counter}')
        
#Plotting histogram for labels
axs[3,0].hist(y,bins=20,color='red',alpha=0.7)
axs[3,0].set_title('Labels')

plt.show()
```

```python
#Calculate the correlation Matrix
cor_matrix=np.corrcoef(X,rowvar=False)

#Plot the correlation matrix
plt.figure(figsize=(9,9))
cax=plt.matshow(cor_matrix,cmap='coolwarm')
plt.colorbar(cax)
plt.title('Correlation Matrix',pad=20)
plt.show()
```

```python
#correlation matrix shows us the redundent data features post removal of Which improves the model accuracy..
#From the Above Observation (0,4) and (2,7) are redundant features..
#We Have Few Threshold Values Which Can be Removed to improve the accuracy..
```

```python
encoder = OneHotEncoder(sparse_output=False)
y_enc = encoder.fit_transform(y.reshape(-1,1))
print(y_enc.shape)
```

```python
#Split training and testing data.. using train_test_split.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y_enc,test_size=0.2,random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Standardizing the features..
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
#Build the neural network model
model=Sequential()

#Add layers to the model
model.add(Dense(32,activation='relu',input_shape=(X_train.shape[1],)))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(len(set(y)),activation='softmax'))
```

```python
#Compile the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

print(model.summary())
```

```python
#Train the model
history=model.fit(X_train_scaled,y_train,epochs=50,batch_size=32,validation_split=0.2)
```

```python
#Visualize Model Training History
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
```

```python
#Summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, y_test)

print("Test Loss:", loss)
print("Test Accuracy:", accuracy)

predictions = model.predict (X_test)
print(predictions.shape, y_test.shape)
```

```python
#Confusion Matrix and Classification report gives the overall views on the test data..
#Precision,Accuracy and loss can be seen in the reports.. 
from sklearn.metrics import confusion_matrix, classification_report

cnf_mtrx = confusion_matrix(np.argmax(y_test, axis=1), np.argmax(predictions, axis=1))
print(cnf_mtrx)

cls_rpt = classification_report(np.argmax(y_test, axis=1), np.argmax(predictions, axis=1))
print(cls_rpt)
```
